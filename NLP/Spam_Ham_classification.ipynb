```python
import pandas as pd

# Read only the first two columns (label and message)
messages = pd.read_csv("C:/Users/HP/Desktop/krish/NLP/spam.csv", encoding='latin-1', usecols=[0, 1])

# Rename columns for clarity
messages.columns = ['label', 'message']
```

```python
messages
```

```python
import re
import nltk 
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
ps = PorterStemmer()
```

```python
corpus = []
for i in range(0 , len(messages)):
    review = re.sub('[^a-zA-Z]', ' ',messages['message'][i]) # remove all non-alphabetic characters
    review = review.lower() # convert to lowercase
    review = review.split() # spilit the string into words
    review = [ps.stem(word) for word in review if not word in stopwords.words('english')] # remove stopwords and stem the words
    review = ' '.join(review) # joining the words back into a single string
    corpus.append(review) # append the cleaned review to the corpus
```

```python
corpus 
```

BAG OF WORDS

```python
y = pd.get_dummies(messages['label']) # dependent variable
y = y.iloc[:,0].values # convert to numpy array
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(corpus,y,test_size = 0.2 )
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=2500,ngram_range=(1,2)) # unigrams and bigrams
```

```python
len(X_train) ,len(y_train)
```

```python
# independent features
X_train = cv.fit_transform(X_train).toarray()  
X_test = cv.transform(X_test).toarray()  
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
spam_detection = MultinomialNB().fit(X_train,y_train) # fit the model on training data
```

```python
y_pred = spam_detection.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
```

```python
accuracy_score(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```

TF-IDF


```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(corpus,y,test_size = 0.2 )
```

```python
## creating the TF-IDF model
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer(max_features=2500,ngram_range=(1,2)) # unigrams and bigrams
X_train = tfidf.fit_transform(X_train).toarray() # create the TF-IDF modeL
X_test = tfidf.transform(X_test).toarray() # transform the test data
```

```python
X_train
```

```python
tfidf.vocabulary_
```

```python
# Fit a new MultinomialNB model on the TF-IDF features if not already done
spam_tfidf = MultinomialNB().fit(X_train, y_train)
y_pred = spam_tfidf.predict(X_test)
```

```python
score = accuracy_score(y_test, y_pred)
print(score)
```

```python
print(classification_report(y_test,y_pred))
```
